```python
cat_cols = ['WindGustDir', 'WindDir9am', 'WindDir3pm']
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
nums_cols = ['Pressure9am','Pressure3pm','WindGustSpeed']
df[nums_cols]= df[nums_cols].fillna(df[nums_cols].median())
```

```python
df = df.dropna(subset=['RainTomorrow'] )
```

```python
df["temp_mean"]= df[["Temp9am","Temp3pm"]].mean(axis=1)
df["windspeed"]= df[["WindSpeed9am","WindSpeed3pm"]].mean(axis=1)
df["humidity_mean"]= df[["Humidity9am","Humidity3pm"]].mean(axis=1)
df["pressure_mean"]= df[["Pressure9am","Pressure3pm"]].mean(axis=1)

df
```

```python
df= df.drop(['Temp9am','Temp3pm','Humidity9am','Humidity3pm','Pressure9am','Pressure3pm','WindSpeed9am','WindSpeed3pm','MinTemp','Location','MaxTemp'],axis ='columns')
```

```python
df.head()
```

```python
df1 = pd.get_dummies(df.RainToday,dtype=int)
df1
```

```python
df3 = pd.get_dummies(df.RainTomorrow,dtype=int)
df3
```

```python
y=df3.drop(['No'],axis=1)
y
```

```python
main = pd.concat([df,df1],axis=1)
main
```

```python
main1 = main.drop(['RainToday','No','RainTomorrow'],axis=1)
main1.head()
```

```python
df2 = pd.get_dummies(
    main1,
    columns=[ "WindGustDir", "WindDir9am", "WindDir3pm"],
    drop_first=True,dtype=int
)
df2
```

```python
print(df2.isnull().sum().sum())
```

```python
df2 = df2.fillna(df2.median())
```

```python
df2.isnull().sum().sort_values(ascending=False)
```

```python
x=df2
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test ,y_train ,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier( n_estimators=100,random_state=42)
model= model.fit(x,y)
```

```python
print(x.shape, y.shape)
print((x.index == y.index).all())
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
